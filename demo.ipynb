# .oppsie Image Format - Interactive Demo

Welcome to the interactive notebook for the **.oppsie** image format! This notebook demonstrates how to load, encode, decode, and convert images using the `.oppsie` reference codec.

```python
# Import required libraries
from PIL import Image
import sys
import os

# Ensure our codec library is in the Python path
sys.path.insert(0, os.path.abspath('.'))
import oppsie
from converter import convert_to_oppsie, convert_from_oppsie, batch_convert
```

## 1. Encoding and Decoding in Memory
Let's create a simple test image using Pillow and encode it to the `.oppsie` format, then decode it back.

```python
# Create a 128x128 gradient image
img = Image.new("RGB", (128, 128))
pixels = []
for y in range(128):
    for x in range(128):
        pixels.append((x * 2, y * 2, (x + y)))
img.putdata(pixels)

# Display original image
print("Original Image:")
display(img)
```

```python
# Encode the image to .oppsie bytes (Lossless)
opps_bytes = oppsie.encode(img)
print(f"Lossless .oppsie Size: {len(opps_bytes)} bytes")

# Decode it back
decoded_img = oppsie.decode(opps_bytes)
print("Decoded Image:")
display(decoded_img)
```

## 2. Lossy Mode (Quantization)
We can apply quantization to lower the color depth, which significantly reduces the `.oppsie` file size for continuous graphics.

```python
# Encode with lossy level 3 (clears lower 3 bits) and level 5
opps_bytes_l3 = oppsie.encode(img, lossy_level=3)
opps_bytes_l5 = oppsie.encode(img, lossy_level=5)

print(f"Lossless size: {len(opps_bytes)} bytes")
print(f"Lossy L3 size: {len(opps_bytes_l3)} bytes (Ratio: {len(opps_bytes_l3)/len(opps_bytes)*100:.1f}%)")
print(f"Lossy L5 size: {len(opps_bytes_l5)} bytes (Ratio: {len(opps_bytes_l5)/len(opps_bytes)*100:.1f}%)")

# Decode and show L5 image
decoded_l5 = oppsie.decode(opps_bytes_l5)
print("Decoded Lossy L5 Image (Notice color banding):")
display(decoded_l5)
```

## 3. Format Conversion CLI Engine
We can also use the converter functions to save directly to disk.

```python
# Save original image as PNG
img.save("temp_original.png")

# Convert PNG to .oppsie
convert_to_oppsie("temp_original.png", "temp_converted.oppsie", lossy_level=3)

# Convert .oppsie back to WebP
convert_from_oppsie("temp_converted.oppsie", "temp_recovered.webp")

print("Files created:")
for f in ["temp_original.png", "temp_converted.oppsie", "temp_recovered.webp"]:
    if os.path.exists(f):
        print(f"  {f:<22} | Size: {os.path.getsize(f)} bytes")
```
